```python
# guardar numeros pares en una lista
# lista

pares_list = []
for number in range(1, 11):
    if (number % 2) == 0:
        pares_list.append(number)

pares_list
```

```python
# guardar en una lista el cuadrado de los numeros impares del 1 al 100

imparescuadr_list = []
for number in range(1, 101):
    if (number % 2) == 1:
        print(f"El cuadrado de este {number} es", number**2)
        imparescuadr_list.append(number**2)
```

```python
imparescuadr_list
```

```python
len(imparescuadr_list)
```

```python
for imparescuadr_list in range(len(imparescuadr_list)):
    print(imparescuadr_list)
```
