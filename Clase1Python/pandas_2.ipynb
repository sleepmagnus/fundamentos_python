```python
# Pasajeros en las clases 2 y 3
df_titanic['Pclass'].unique()
```

```python
# pasajeros pertenecientes a la clase 2 y 3
df_class23 = df_titanic[df_titanic['Pclass'].isin([2,3])]
```

```python
# passajeros pertenecientes a la clase 3
df_class3 = df_titanic[df_titanic['Pclass'].isin([3])]
```

```python
df_class23.shape
```

```python
# porcentaje de pasajeros que eran pertenecientes a la clase 3
df_class3.shape[0] * 100 / df_titanic.shape[0]
```

```python
# pasajeros pertenecientes a la clase 2 y 3 con el operador or en pandas
# el operador es este (|) se encuentra a la izquierda del 1 con el teclado en español
df_c23 = df_titanic[ (df_titanic['Pclass'] == 2) | (df_titanic['Pclass'] == 3)]
```

```python
df_c23.info()
```

```python
# pasajeros de las clases 2 y 3 con edad valida
# notna me muestra las filas con valores validos
df_c23_age = df_c23[ df_c23['Age'].notna()]
```

```python
# primero inspeccionas, despues metes lo que hiciste en una variable para mas facilidad
df_c23_age
```
