```python
# funciones
def suma_dos(num1:int, num2:int):
    """
    Esta funcion suma dos numeros enteros
    num1: int
    num2: int
    return: int
    """
    resultado = num1 + num2
    return num1 + num2
```

```python
# base de argumentos por posicion
suma_dos(2, 4)
```

```python
# funciones
def resta_dos(num1:int, num2:int):
    """
    Esta funcion resta num1 - num2
    num1: int
    num2: int
    return: int
    """
    resultado = num1 - num2
    return resultado
```

```python
# base de argumentos por parametros
resta_dos(num2=9, num1=8)
```

```python
# funciones
def multi_dos_positivos(num1, num2):
    """
    Esta funcion multiplica num1 - num2
    num1: int
    num2: int
    return: int
    """
    if (num1 > 0) and (num2 > 0):
        resultado = num1 * num2
    else:
        print("num1 o num2 es menor que cero, revisar")
    return resultado
```

```python
multi_dos_positivos (4, 3)
```
